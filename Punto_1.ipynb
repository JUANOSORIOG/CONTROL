```python
import numpy as np
import control.matlab as ctrl
import matplotlib.pyplot as plt
# Definición de la función de transferencia G1(s)
num1 = [30]
den1 = [1, 30]
G1 = ctrl.tf(num1, den1)

# Cálculo y gráfica del Diagrama de Bode para G1
plt.figure(figsize=(10, 8))
# Usamos ctrl.bode() para obtener las trazas (mag en lineal, fase en grados)
mag_lin1, phase_rad1, omega1 = ctrl.bode(G1, plot=False)

# Graficamos con las utilidades de matplotlib
# Magnitud en dB
plt.subplot(2, 1, 1)
mag_dB1 = 20 * np.log10(mag_lin1) # Conversión a dB
plt.semilogx(omega1, mag_dB1)
plt.title('Diagrama de Bode para $G_1(s) = \\frac{30}{s+30}$')
plt.ylabel('Magnitud (dB)')
plt.grid(which="both")

# Fase en grados
plt.subplot(2, 1, 2)
phase_deg1 = np.degrees(phase_rad1) # Conversión a grados
plt.semilogx(omega1, phase_deg1)
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Fase (grados)')
plt.grid(which="both")
plt.show()
```

```python
w_a = 5 # Frecuencia de la entrada
U_a = np.sqrt(2) # Amplitud de la entrada
j_w_a = 1j * w_a

# Evaluar G1(j*omega)
G1_jw_a = ctrl.evalfr(G1, j_w_a)

# Magnitud y Fase de G1(j*omega)
mag_G1_a = np.abs(G1_jw_a)
phase_G1_a = np.angle(G1_jw_a) # en radianes

# Salida en estado estable
Y_a = U_a * mag_G1_a
phi_a = phase_G1_a

# Vector de tiempo
t_a = np.linspace(0, 5, 500) # Simulación por 5 segundos

# Señal de entrada
u_t_a = U_a * np.cos(w_a * t_a)

# Señal de salida en estado estable
y_t_a = Y_a * np.cos(w_a * t_a + phi_a)

# Graficar
plt.figure(figsize=(10, 5))
plt.plot(t_a, u_t_a, label=f'Entrada $u(t) = {U_a:.2f}\\cos({w_a}t)$')
plt.plot(t_a, y_t_a, label=f'Salida $y(t) = {Y_a:.2f}\\cos({w_a}t + {np.degrees(phi_a):.2f}°)$')
plt.title(f'Respuesta en Frecuencia de $G_1(s)$ para $\\omega = {w_a}$ rad/s')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.show()
```
