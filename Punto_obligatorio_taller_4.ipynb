Ahora, importaremos las librerías necesarias, definiremos una función de transferencia de ejemplo y un punto de prueba en el plano complejo.

```python
import control as ct
import numpy as np

# Definir una función de transferencia de ejemplo
# Por ejemplo, G(s) = (s + 2) / (s^2 + 2s + 1)
num = [1, 2]
den = [1, 2, 1]
G = ct.tf(num, den)

print(f"Función de Transferencia: {G}")

# Definir un punto de prueba en el plano complejo (por ejemplo, s = -1 + 1j)
test_point = -1 + 1j
print(f"Punto de Prueba (s): {test_point}")
```

Para calcular la magnitud, el ángulo y la ganancia, evaluamos la función de transferencia en el punto de prueba `s`. La 'ganancia' en este contexto se refiere comúnmente a la magnitud del valor de la función de transferencia en ese punto complejo.

```python
# Evaluar la función de transferencia en el punto de prueba
# La forma correcta de evaluar G(s) en un punto complejo s es G(s)
result = G(test_point)

# Calcular magnitud
magnitud = np.abs(result)

# Calcular ángulo (en grados)
angulo = np.angle(result, deg=True)

# La ganancia se considera la magnitud en este contexto
ganancia = magnitud

print(f"Valor de G(s) en el punto de prueba: {result:.4f}")
print(f"Magnitud de G(s): {magnitud:.4f}")
print(f"Ángulo de G(s): {angulo:.2f} grados")
print(f"Ganancia (Magnitud) de G(s): {ganancia:.4f}")
```
