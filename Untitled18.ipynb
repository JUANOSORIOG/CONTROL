```python
import control.matlab as ctrl
import numpy as np
import matplotlib.pyplot as plt

# 1. Definición de la Función de Transferencia G(s) (Lazo Abierto)
# G(s) = 0.3188 / (s^2 + 1.1435s + 0.3188)
num_G = [0.3188]
den_G = [1, 1.1435, 0.3188]
G_tanques = ctrl.tf(num_G, den_G)

print("Función de Transferencia en Lazo Abierto G(s):\n", G_tanques)
print("-" * 60)

# 2. Definición de la Función de Transferencia T(s) (Lazo Cerrado)
# T(s) = 0.3188 / (s^2 + 1.1435s + 0.6373)
num_T = [0.3188]
den_T = [1, 1.1435, 0.6373]
T_tanques = ctrl.tf(num_T, den_T)

print("Función de Transferencia en Lazo Cerrado T(s):\n", T_tanques)
print("-" * 60)

# 3. Cálculo de Márgenes y Frecuencias de Cruce para G(s)
# gm (Margen de Ganancia - unidades absolutas)
# pm (Margen de Fase - grados)
gm, pm, wcg, wcp = ctrl.margin(G_tanques)

# Convertir Margen de Ganancia a decibeles (dB)
gm_dB = 20 * np.log10(gm)

# 4. Impresión de Resultados de Estabilidad para G(s)
print("Resultados de Estabilidad (Márgenes de G(s)):")
print(f"  > Margen de Ganancia (GM): {gm_dB:.2f} dB (Infinito teórico)")
print(f"  > Margen de Fase (PM): {pm:.2f} grados")
print("-" * 60)


# 5. Generación y Graficación de los Diagramas de Bode
omega = np.logspace(-3, 3, 500)

# Bode para G(s) (Lazo Abierto)
mag_G, phase_G, omega_G = ctrl.bode(G_tanques, omega)
mag_G_dB = 20 * np.log10(mag_G)

# Bode para T(s) (Lazo Cerrado)
mag_T, phase_T, omega_T = ctrl.bode(T_tanques, omega)
mag_T_dB = 20 * np.log10(mag_T)


plt.figure(figsize=(12, 10))

# Gráfico de Magnitud para G(s) y T(s)
plt.subplot(2, 1, 1)
plt.semilogx(omega_G, mag_G_dB, color='blue', label='$G(s)$ (Lazo Abierto)')
plt.semilogx(omega_T, mag_T_dB, color='green', label='$T(s)$ (Lazo Cerrado)')
plt.title('Diagrama de Bode de las Funciones de Transferencia')
plt.ylabel('Magnitud (dB)')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # 0 dB
plt.legend()

# Gráfico de Fase para G(s) y T(s)
plt.subplot(2, 1, 2)
plt.semilogx(omega_G, phase_G * 180 / np.pi, color='blue', label='$G(s)$ (Lazo Abierto)') # Convert phase to degrees
plt.semilogx(omega_T, phase_T * 180 / np.pi, color='green', label='$T(s)$ (Lazo Cerrado)') # Convert phase to degrees
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Fase (grados)')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.axhline(-180, color='black', linestyle='--', linewidth=0.8) # -180 grados
plt.legend()

plt.tight_layout()
plt.show()
```

## Definir la función de transferencia para k=10

### Subtask:
Crear el objeto de función de transferencia para $G(s) = \frac{K}{s^3 + 6s^2 + 12s}$ con $K=10$.


**Reasoning**:
Define the numerator and denominator coefficients for the transfer function G(s) with K=10 and create the transfer function object.



```python
# Define los coeficientes del numerador y denominador para K=10
K = 10
num_G_k10 = [K]
den_G = [1, 6, 12, 0] # Coeficientes de s^3 + 6s^2 + 12s

# Crea el objeto de función de transferencia
G_k10 = ctrl.tf(num_G_k10, den_G)

# Imprime la función de transferencia
print("Función de Transferencia G(s) con K=10:\n", G_k10)
```

## Graficar bode para k=10

### Subtask:
Generar y mostrar el diagrama de Bode (magnitud y fase) para $G(s)$ con $K=10$.


**Reasoning**:
Generate and plot the Bode diagram for G(s) with K=10, including magnitude in dB and phase in degrees, following the specified plotting requirements.



```python
# Generate frequency range
omega = np.logspace(-2, 2, 400)

# Generate Bode plot data for G_k10
mag_G_k10, phase_G_k10, omega_G_k10 = ctrl.bode(G_k10, omega)

# Convert magnitude to dB
mag_G_k10_dB = 20 * np.log10(mag_G_k10)

# Convert phase to degrees
phase_G_k10_deg = phase_G_k10 * 180 / np.pi

# Create figure and subplots
plt.figure(figsize=(10, 8))

# Magnitude plot
plt.subplot(2, 1, 1)
plt.semilogx(omega_G_k10, mag_G_k10_dB, color='blue')
plt.title('Diagrama de Bode de G(s) con K=10')
plt.ylabel('Magnitud (dB)')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # 0 dB line

# Phase plot
plt.subplot(2, 1, 2)
plt.semilogx(omega_G_k10, phase_G_k10_deg, color='red')
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Fase (grados)')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.axhline(-180, color='black', linestyle='--', linewidth=0.8) # -180 degrees line

plt.tight_layout()
plt.show()
```

## Calcular márgenes de estabilidad para k=10

### Subtask:
Calcular el margen de ganancia y el margen de fase para $G(s)$ con $K=10$.


**Reasoning**:
Calculate the stability margins for the transfer function `G_k10` and print the results.



```python
# Calculate stability margins for G_k10
gm, pm, wcg, wcp = ctrl.margin(G_k10)

# Convert Gain Margin to dB
gm_dB = 20 * np.log10(gm)

# Print the calculated margins and crossover frequencies
print("Resultados de Estabilidad para G(s) con K=10:")
print(f"  > Margen de Ganancia (GM): {gm_dB:.2f} dB")
print(f"  > Margen de Fase (PM): {pm:.2f} grados")
print(f"  > Frecuencia de Cruce de Ganancia (wcg): {wcg:.2f} rad/s")
print(f"  > Frecuencia de Cruce de Fase (wcp): {wcp:.2f} rad/s")
```

## Graficar bode para k=80

### Subtask:
Generar y mostrar el diagrama de Bode (magnitud y fase) para $G(s)$ con $K=80$.

**Reasoning**:
Generate Bode plot data for G(s) with K=80, convert the magnitude to dB and phase to degrees, then plot the Bode diagram, including titles, labels, grids, and reference lines.

```python
# Define los coeficientes del numerador y denominador para K=80
K = 80
num_G_k80 = [K]
den_G = [1, 6, 12, 0] # Coeficientes de s^3 + 6s^2 + 12s

# Crea el objeto de función de transferencia
G_k80 = ctrl.tf(num_G_k80, den_G)

# Generate frequency range
omega = np.logspace(-2, 2, 400)

# Generate Bode plot data for G_k80
mag_G_k80, phase_G_k80, omega_G_k80 = ctrl.bode(G_k80, omega)

# Convert magnitude to dB
mag_G_k80_dB = 20 * np.log10(mag_G_k80)

# Convert phase to degrees
phase_G_k80_deg = phase_G_k80 * 180 / np.pi

# Create figure and subplots
plt.figure(figsize=(10, 8))

# Magnitude plot
plt.subplot(2, 1, 1)
plt.semilogx(omega_G_k80, mag_G_k80_dB, color='red')
plt.title('Diagrama de Bode de G(s) con K=80')
plt.ylabel('Magnitud (dB)')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # 0 dB line

# Phase plot
plt.subplot(2, 1, 2)
plt.semilogx(omega_G_k80, phase_G_k80_deg, color='red')
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Fase (grados)')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.axhline(-180, color='black', linestyle='--', linewidth=0.8) # -180 degrees line

plt.tight_layout()
plt.show()
```

## Summary:

### Data Analysis Key Findings

*   For K=10, the system has a Gain Margin (GM) of 17.15 dB and a Phase Margin (PM) of 66.80 degrees, indicating a stable system with good relative stability.
*   For K=80, the system has a Gain Margin (GM) of -0.92 dB and a Phase Margin (PM) of -3.42 degrees, indicating an unstable system in closed loop.
*   Increasing the gain K from 10 to 80 causes both the gain and phase margins to decrease significantly and become negative.
*   Visually, increasing K shifts the magnitude plot upwards, causing the 0 dB crossover frequency to decrease. The phase plot remains unchanged, but the decreased 0 dB crossover frequency means the phase at that frequency is closer to or beyond -180 degrees.

### Insights or Next Steps

*   The system stability is highly sensitive to the gain K. There is a critical value of K between 10 and 80 where the system transitions from stable to unstable.
*   Future steps could involve finding the exact value of K for which the system becomes marginally stable (GM or PM equals 0) to determine the stability limit.
