```python
import control.matlab as ctrl
import numpy as np
import matplotlib.pyplot as plt

# 1. Definición de la Función de Transferencia G(s) (Lazo Abierto)
# G(s) = 0.3188 / (s^2 + 1.1435s + 0.3188)
num_G = [0.3188]
den_G = [1, 1.1435, 0.3188]
G_tanques = ctrl.tf(num_G, den_G)

print("Función de Transferencia en Lazo Abierto G(s):\n", G_tanques)
print("-" * 60)

# 2. Definición de la Función de Transferencia T(s) (Lazo Cerrado)
# T(s) = 0.3188 / (s^2 + 1.1435s + 0.6373)
num_T = [0.3188]
den_T = [1, 1.1435, 0.6373]
T_tanques = ctrl.tf(num_T, den_T)

print("Función de Transferencia en Lazo Cerrado T(s):\n", T_tanques)
print("-" * 60)

# 3. Cálculo de Márgenes y Frecuencias de Cruce para G(s)
# gm (Margen de Ganancia - unidades absolutas)
# pm (Margen de Fase - grados)
gm, pm, wcg, wcp = ctrl.margin(G_tanques)

# Convertir Margen de Ganancia a decibeles (dB)
gm_dB = 20 * np.log10(gm)

# 4. Impresión de Resultados de Estabilidad para G(s)
print("Resultados de Estabilidad (Márgenes de G(s)):")
print(f"  > Margen de Ganancia (GM): {gm_dB:.2f} dB (Infinito teórico)")
print(f"  > Margen de Fase (PM): {pm:.2f} grados")
print("-" * 60)


# 5. Generación y Graficación de los Diagramas de Bode
omega = np.logspace(-3, 3, 500)

# Bode para G(s) (Lazo Abierto)
mag_G, phase_G, omega_G = ctrl.bode(G_tanques, omega)
mag_G_dB = 20 * np.log10(mag_G)

# Bode para T(s) (Lazo Cerrado)
mag_T, phase_T, omega_T = ctrl.bode(T_tanques, omega)
mag_T_dB = 20 * np.log10(mag_T)


plt.figure(figsize=(12, 10))

# Gráfico de Magnitud para G(s) y T(s)
plt.subplot(2, 1, 1)
plt.semilogx(omega_G, mag_G_dB, color='blue', label='$G(s)$ (Lazo Abierto)')
plt.semilogx(omega_T, mag_T_dB, color='green', label='$T(s)$ (Lazo Cerrado)')
plt.title('Diagrama de Bode de las Funciones de Transferencia')
plt.ylabel('Magnitud (dB)')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.axhline(0, color='black', linestyle='--', linewidth=0.8) # 0 dB
plt.legend()

# Gráfico de Fase para G(s) y T(s)
plt.subplot(2, 1, 2)
plt.semilogx(omega_G, phase_G * 180 / np.pi, color='blue', label='$G(s)$ (Lazo Abierto)') # Convert phase to degrees
plt.semilogx(omega_T, phase_T * 180 / np.pi, color='green', label='$T(s)$ (Lazo Cerrado)') # Convert phase to degrees
plt.xlabel('Frecuencia (rad/s)')
plt.ylabel('Fase (grados)')
plt.grid(True, which="both", ls="-", alpha=0.6)
plt.axhline(-180, color='black', linestyle='--', linewidth=0.8) # -180 grados
plt.legend()

plt.tight_layout()
plt.show()
```
